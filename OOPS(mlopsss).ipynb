```python
# Step 1: Import Required Libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report


# Step 2: Data Loader Class
class DataLoader:
    def __init__(self):
        self.X, self.y = load_iris(return_X_y=True)


    def split(self, test_size=0.3, random_state=42):
        return train_test_split(self.X, self.y, test_size=test_size, random_state=random_state)


# Step 3: Preprocessor Class
class Preprocessor:
    def __init__(self):
        self.scaler = StandardScaler()


    def fit_transform(self, X_train):
        return self.scaler.fit_transform(X_train)


    def transform(self, X_test):
        return self.scaler.transform(X_test)


# Step 4: ML Model Class
class MLModel:
    def __init__(self):
        self.model = DecisionTreeClassifier()


    def train(self, X_train, y_train):
        self.model.fit(X_train, y_train)


    def predict(self, X_test):
        return self.model.predict(X_test)


# Step 5: Evaluator Class
class Evaluator:
    def __init__(self, y_true, y_pred):
        self.y_true = y_true
        self.y_pred = y_pred


    def report(self):
        print("ðŸ“Š Classification Report:\n")

        print(classification_report(self.y_true, self.y_pred))


# Step 6: Main ML Application Class
class MLApplication:
    def __init__(self):
        self.loader = DataLoader()
        self.preprocessor = Preprocessor()
        self.model = MLModel()


    def run(self):
        # Load and split data
        X_train, X_test, y_train, y_test = self.loader.split()

        # Preprocess data
        X_train_scaled = self.preprocessor.fit_transform(X_train)
        X_test_scaled = self.preprocessor.transform(X_test)


        # Train model
        self.model.train(X_train_scaled, y_train)
        y_pred = self.model.predict(X_test_scaled)


        # Evaluate
        evaluator = Evaluator(y_test, y_pred)
        evaluator.report()


# Step 7: Execute the Pipeline
app = MLApplication()
app.run()



```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Base class for loading data
class DataLoader:
    def __init__(self):
        self.X, self.y = load_iris(return_X_y=True)

    def split(self, test_size=0.3, random_state=42):
        return train_test_split(self.X, self.y, test_size=test_size, random_state=random_state)

# Base class for preprocessing
class Preprocessor:
    def __init__(self):
        self.scaler = StandardScaler()

    def fit_transform(self, X_train):
        return self.scaler.fit_transform(X_train)

    def transform(self, X_test):
        return self.scaler.transform(X_test)

# Base MLModel class (can be inherited by different models)
class MLModel:
    def __init__(self):
        self.model = None  # Placeholder to be defined in child

    def train(self, X_train, y_train):
        self.model.fit(X_train, y_train)

    def predict(self, X_test):
        return self.model.predict(X_test)

# Derived class using SVM
class SVMModel(MLModel):
    def __init__(self):
        super().__init__()
        self.model = SVC(kernel='linear')

# Evaluation class
class Evaluator:
    def __init__(self, y_true, y_pred):
        self.y_true = y_true
        self.y_pred = y_pred

    def report(self):
        print("Classification Report:\n")
        print(classification_report(self.y_true, self.y_pred))

# Main ML pipeline application class
class MLApplication(DataLoader, Preprocessor, SVMModel):
    def __init__(self):
        DataLoader.__init__(self)
        Preprocessor.__init__(self)
        SVMModel.__init__(self)

    def run(self):
        # Load and split data
        X_train, X_test, y_train, y_test = self.split()

        # Preprocess data
        X_train_scaled = self.fit_transform(X_train)
        X_test_scaled = self.transform(X_test)

        # Train model
        self.train(X_train_scaled, y_train)
        y_pred = self.predict(X_test_scaled)

        # Evaluate
        evaluator = Evaluator(y_test, y_pred)
        evaluator.report()

# Run the application
app = MLApplication()
app.run()
```
